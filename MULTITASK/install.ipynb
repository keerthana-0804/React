```python
import streamlit as st
import subprocess

def run_script():
    # Call your Python script here
    subprocess.run(["python", "script.py"], capture_output=True)

# Streamlit UI
st.title("Sample UI")

# Define UI elements
st.write("Click the button below to run the script:")
if st.button("Run Script"):
    run_script()
```
